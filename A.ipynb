```python
def window(data, x, y, half_size):
    return data[x - half_size:x + half_size, y - half_size:y + half_size]

def apply_f(f, data, half_size):
    result = np.zeros(data.shape)
    for i in range(half_size, data.shape[0] - half_size):
        for j in range(half_size, data.shape[1] - half_size):
            try:
                result[i][j] = f(window(data, i, j, half_size))
            except Exception as e:
                print e
                print i, j
                return None
                
    return result
```

```python
averager = lambda x: int(sum(x) / x.size)
averages = apply_f(averager, arr, 5)
```

```python
thresholder = lambda t: lambda x: 255 if averager(x) > t else 0
threshold_light = apply_f(thresholder(190), arr, 1)
```

```python
thresholder_inverse = lambda t: lambda x: 255 if averager(x) < t else 0
threshold_dark = apply_f(thresholder_inverse(80), arr, 1)
```

```python
# Sparse Serialize
def sparse(f, data):
    result = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            try:
                if (data[i, j] > 0):
                    result.append("x\n%s\ny\n%s\nf\n%s" % (i, j, data[i, j]))
            except Exception as e:
                print e
                print i, j
                print X, Y
                return None
            
    print >> f, "w\n%s" % data.shape[0]
    print >> f, "h\n%s" % data.shape[1]
    print >> f, "l\n%s" % len(result)
    print >> f, "\n".join(result)

# Dense Serialize
def dense(f, data):
    result = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            try:
                if (data[i, j] > 0):
                    result.append("%s" % (data[i, j]))
            except Exception as e:
                print e
                print i, j
                print X, Y
                return None
            
    print >> f, "w\n%s" % data.shape[0]
    print >> f, "h\n%s" % data.shape[1]
    print >> f, "\n".join(result)


with open('analysis_sparse.yolo', 'w') as f:
    print >> f, "raw(dense)"
    dense(f, arr)
    print >> f, "burning(sparse)"
    sparse(f, threshold_light)
    print >> f, "freezing(sparse)"
    sparse(f, threshold_dark)
    print >> f, "skylight(sparse)"
    sparse(f, skylight_proba)
    
    
    
```
